```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
digits =  datasets.load_digits()

plt.imshow(digits.images[0], cmap='gray')
```

```python
def plot_multi():
  nplots = 16
  fig = plt.figure(figsize=(15,15))
  for j in range(nplots):
    plt.subplot(4,4,j+1)
    plt.imshow(digits.images[j], cmap='binary')
    plt.title(digits.target[j])
    plt.axis('off')
plot_multi()
```

```python
y = digits.target
x = digits.images
```

```python
x = x.reshape(len(x), -1)
```

```python
x_train = x[:1000]
y_train = y[:1000]
x_test = x[1000:]
y_test = y[1000:]
```

```python
from sklearn.neural_network import MLPClassifier

# Créez un nouveau modèle MLPClassifier avec les paramètres spécifiés
model = MLPClassifier(
    hidden_layer_sizes=(15,),  # Taille de la couche cachée, une couche de 15 neurones
    activation='logistic',     # Fonction d'activation logistique
    solver='sgd',             # Solveur de descente de gradient stochastique
    tol=1e-4,                 # Tolérance pour la convergence
    random_state=1,           # Graine aléatoire pour la reproductibilité
    learning_rate_init=0.1,   # Taux d'apprentissage initial
    verbose=True              # Affichez les messages de progression
)
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
print (y_pred)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy Score:", accuracy)
```
